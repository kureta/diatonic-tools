```python
from diatonic_tools.init import plt, sns
import numpy as np
```

```python
def unit_circle():
    theta = np.linspace(0, 2*np.pi, 100)
    x = np.cos(theta)
    y = np.sin(theta)

    return x, y
```

```python
step = np.exp(2j*np.pi/12)
```

```python
def plot_complex_number(z):
    plt.plot(z.real, z.imag, 'ro')
    plt.plot([0, z.real], [0, z.imag])
```

```python
def decorate_plot(f):
    def f_():
        plt.plot(*unit_circle())
    
        f()
    
        plt.xlim(-1.1, 1.1)
        plt.ylim(-1.1, 1.1)
        plt.gca().set_aspect('equal', adjustable='box')
        plt.grid(True)
        plt.show()
        plt.close()

    return f_
```

```python
major_triad = [step**0, step**4, step**7]
minor_triad = [step**0, step**3, step**7]
```

```python
for i in range(8, 12):
    chord = major_triad + [step**i]
    @decorate_plot
    def balance():
        for c in chord:
            plot_complex_number(c)
        plot_complex_number(np.sum(chord))

    balance()
```

```python
for i in range(8, 12):
    chord = minor_triad + [step**i]
    @decorate_plot
    def balance():
        for c in chord:
            plot_complex_number(c)
        plot_complex_number(np.sum(chord))

    balance()
```
